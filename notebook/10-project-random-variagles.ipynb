```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, binom, uniform


np.random.seed(42)
```

## Binomial distribution

En teoría de la probabilidad y estadística, la distribución binomial es la distribución de probabilidad discreta que da sólo dos resultados posibles en un experimento; por ejemplo, al lanzar una moneda, puede salir cara o cruz.

```python
def puntoestimado_distbinomial(x, size, prob=0.5):
    """
    Calculates the point estimate of the binomial distribution
    """
    result = binom.pmf(k=x, n=size, p=prob, loc=0)
    return result


def acumulado_distbinomial(q, size, prob=0.5):
    """
    Calculates the cumulative of the binomial distribution
    """
    result = binom.cdf(k=q, n=size, p=prob, loc=0)
    return result


def cuantil_distbinomial(p, size, prob=0.5):
    """
    Calculates the quantile function from the binomial distribution
    """
    result = binom.ppf(q=p, n=size, p=prob, loc=0)
    return result


def variables_aleatorias(n, size, prob=0.5):
    """
    Generates random variables from the binomial distribution
    """
    result = binom.rvs(n=size, p=prob, size=n)
    return result
```

```python
data = variables_aleatorias(2000, size=10, prob=0.2)
pmf = puntoestimado_distbinomial(data, size=10, prob=0.2)
cdf = acumulado_distbinomial(data, size=10, prob=0.2)
ppf = cuantil_distbinomial(data, size=10, prob=0.2)
```

```python
fig, axis = plt.subplots(3, 1, figsize=(5, 7))
axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

plt.tight_layout()  # Adjust the layout
plt.show()  # Show the plot
```

### Distribuciones normales

```python
def puntoestimado_distbinomial(x, mean=0, std=1):
    result = norm.pdf(x, loc=mean, scale=std)
    return result


def acumulado_distbinomial(x, mean = 0, std = 1):
    result = norm.cdf(x = x, loc = mean, scale = std)
    return result


def cuantil_distbinomial(p, mean = 0, std = 1):
    result = norm.ppf(q = p, loc = mean, scale = std)
    return result


def variables_aleatorias(n, mean = 0, std = 1):
    result = norm.rvs(size = n, loc = mean, scale = std)
    return result
```

```python
mean = 0
std = 1
data = np.arange(-5, 5, 0.01)


pdf = puntoestimado_distbinomial(data, mean=mean, std=std)
cdf = acumulado_distbinomial(data, mean=mean, std=std)
ppf = cuantil_distbinomial(data, mean=mean, std=std)
```

```python
fig, axis = plt.subplots(3, 1, figsize =(5, 7))


axis[0].plot(data, pdf, "r-")
axis[0].set_title('Punto estimado de la distribución binomial')
axis[1].plot(data, cdf, "b-")
axis[1].set_title('Acumulado de la distribución binomial')
axis[2].plot(data, ppf, "g-")
axis[2].set_title('Cuantil de la distribución binomial')


plt.tight_layout()
plt.show()
```

### Distribuciones Uniformes

```python
def dunif(x, low=0, high=1):
    result = uniform.pdf(x, loc = low, scale = (high - low))
    return result


def punif(q, low=0, high=1):
    result = uniform.cdf(q, loc = low, scale = (high - low))
    return result


def qunif(p, low=0, high=1):
    result = uniform.ppf(p, loc = low, scale = (high - low))
    return result


def runif(n, low=0, high=1):
    result = uniform.rvs(loc = low, scale = (high - low), size = n)
    return result
```

```python
low = 0
high = 1
data = np.arange(-5, 5, 0.01)


pmf = dunif(data, low=low, high=high)
cdf = punif(data, low=low, high=high)
ppf = qunif(data, low=low, high=high)
```

```python
fig, axis = plt.subplots(3, 1, figsize=(5, 7))


axis[0].plot(data, pmf, "r-")
axis[0].set_title('Punto estimado de la distribución binomial')
axis[1].plot(data, cdf, "b-")
axis[1].set_title('Acumulado de la distribución binomial')
axis[2].plot(data, ppf, "g-")
axis[2].set_title('Cuantil de la distribución binomial')

plt.tight_layout()
plt.show()
```
